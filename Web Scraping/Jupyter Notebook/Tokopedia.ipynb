```python
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import Select


driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.tokopedia.com')

keyword = driver.find_element_by_xpath('//*[@id="search-container"]/form/div/div/input')
keyword.send_keys("Kue Lapis Legit")
search = driver.find_element_by_xpath('//*[@id="search-container"]/form/div/div/button')
search.click()
```

```python
driver.find_element_by_class_name('css-1g20a2m').text
```

```python
search = driver.find_element_by_xpath('//*[@id="search-container"]/form/div/div/button')
search.click()
i = 1
listoflinks = []
while i < 7:
    listofproducts = driver.find_elements_by_class_name('css-1g20a2m')
    print("Page", i)
    for el in listofproducts:
        lk = el.find_element_by_tag_name('a')
        listoflinks.append(lk.get_property('href'))
    klik = driver.find_element_by_class_name('css-98hn3t.e19tp72t1')
    print('    Next Page >>')
    klik.click()
    i += 1
print('Completed!')
```

```python
import requests
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
from scrapy import Selector
import pandas as pd
from IPython.core.display import HTML
from io import BytesIO
import xlsxwriter
import time
from tqdm import tqdm


headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'}


lapislegit=[]
lisofimage = []
for i in tqdm(listoflinks):
    r = requests.get(i, headers=headers)
    soup = BeautifulSoup(r.content,'lxml')
    productname = soup.find('h1', class_='css-x7lc0h').text
    price = soup.find('h3', class_='css-c820vl').text
    soup.find_element_by_css_selector
    try:
        productratingstar =   Selector(text = requests.get(i, headers=headers).content).xpath('//span[@data-testid="lblPDPDetailProductRatingNumber"]/text()').extract()[0]
    except:
        productratingstar = 0.0
    store = soup.find(class_='css-xmjuvc').text
    storelocation =  Selector(text = requests.get(i, headers=headers).content).xpath('//*[@id="zeus-root"]/div/div[2]/div[2]/div/div/div[1]/div/div[2]/div/span/text()[1]').extract()[0]
    img = Selector(text = requests.get(i, headers=headers).content).xpath('//img[@class="success fade"]/@src').extract()[0]
    
    details={'Product Name':productname,
             'Price':price,
             'Product Rating Star':productratingstar,
             'Store Name':store,
             'Store Location':storelocation}
    
    lapislegit.append(details)
    lisofimage.append(img)
```

```python
import pandas as pd

df = pd.DataFrame(lapislegit)

df['Price'] = df['Price'].str.replace('Rp','')
df['Price'] = df['Price'].str.replace('.','')
df['Price'] = df['Price'].astype(str).astype(int)
df['Product Rating Star'] = df['Product Rating Star'].astype(float)
```

```python
writer = pd.ExcelWriter('Lapislegit Tokopedia.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df.to_excel(writer, sheet_name='Sheet1')

workbook  = writer.book
worksheet = writer.sheets['Sheet1']

worksheet.set_column('G2:G2',12)
worksheet.set_default_row(60)

cell_format = workbook.add_format({'bold': True, 'align': 'center', 'valign': 'top', 'border': 1})
worksheet.write(0,6,'Picture', cell_format)

#insert images
image_row = 1
image_col = 6
try:
    for image in lisofimage:
        image_data = BytesIO(urlopen(image).read())
        worksheet.insert_image(image_row,
                               image_col,
                               image,
                               {'image_data' : image_data, 'x_scale': 0.35, 'y_scale' : 0.35, 'x_offset ': 3.5, 'y_offset' : 3.5})

        #positioning = 1 allows move and size with cells
        image_row += 1
except:
    pass

writer.save()
```
