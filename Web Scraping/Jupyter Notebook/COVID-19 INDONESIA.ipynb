```python
df['Last Update']=pd.to_datetime('today').strftime("%d-%m-%Y")
df
```

```python
Positif_new = (round(df.loc[0:8, ['Positif']] * 1000)).append(df.loc[9:, ['Positif']])
Positif_new
```

```python
Sembuh_new = (round(df.loc[0:3,['Sembuh']] * 1000)).append(df.loc[4:,['Sembuh']])
Sembuh_new
```

```python
dfnew = df.copy()
dfnew.head()
```

```python
dfnew = df.copy()
dfnew['Positif']= Positif_new
dfnew['Sembuh'] = Sembuh_new
dfnew
```

```python
cols = ['Positif','Sembuh']
dfnew[cols] = dfnew[cols].applymap(np.int64)
dfnew
```

```python
dfnew.head()
```

```python
import docx
import pandas as pd

doc = docx.Document()

#heading
doc.add_heading('Data COVID-19 Seluruh Indonesia', level=2)
doc.add_paragraph('These are the list of the price of golds')#, style='Intense Quote')

table = doc.add_table(dfnew.shape[0]+1, dfnew.shape[1])

# header table
for j in range(dfnew.shape[-1]):
    table.cell(0,j).text = dfnew.columns[j]
    
for i in range(dfnew.shape[0]):
    for j in range(dfnew.shape[-1]):
        table.cell(i+1,j).text = str(dfnew.values[i,j])

table.style = 'DarkList'
doc.save('Data Covid-19.docx')
print('File Saved!!')
```

# Existing Table

```python
from docx import Document
import pandas as pd
document = Document('Data Covid-19 Update 2020-06-17.docx')

tables = []
for table in document.tables:
    df = [['' for i in range(len(table.columns))] for j in range(len(table.rows))]
    for i, row in enumerate(table.rows):
        for j, cell in enumerate(row.cells):
            if cell.text:
                df[i][j] = cell.text
    tables.append(pd.DataFrame(df))
print(tables)
```

```python
covid = tables[0]

#Add header as label
header = covid.iloc[0]
covid =covid[1:]
covid.columns = header
covid.head()
```

```python
covid.info()
```

```python
covid[['Positif', 'Sembuh','Meninggal']] = covid[['Positif', 'Sembuh', 'Meninggal']].astype(str).astype(int)
```

```python
covid.info()
```

```python
def color_negative_red(value):
  if value >=2000:
    color = 'red'
  elif value >=1000:
    color = 'yellow'
  elif value >=500:
    color = 'green'
  else:
    color = 'none'

  return 'background-color: %s' % color
```

cov_color = covid[['Positif', 'Sembuh', 'Meninggal']]
cov_color.style.applymap(color_negative_red)

```python
covid = covid.style.applymap(color_negative_red, subset=['Positif', 'Sembuh', 'Meninggal'])
covid
```

```python
covid.to_excel("Data Covid-19 With Color Format.xlsx") 
print('File Saved!!')
```
